## Linear Regression: Model Representation

Implementing the model [f(w, b)] for linear regression with one variable

```python
# Importing the dependencies
import numpy as np  # for scientific computation
import matplotlib.pyplot as plt  # for plotting data
```

```python
# Creating dataset for house price prediction. The dataset constitutes of two points wherein the units for size of the house is in 1000s of sqft.
# and price of the house is in 1000s of dollars
x_train = np.array([1.0, 2.0])  # x_train is the input variable (size in 1000 square feet)
y_train = np.array([300.0, 500.0])  # y_train is the target (price in 1000s of dollars)
print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

```python
# m is the number of training examples
print(f"x_train.shape: {x_train.shape}")  # x_train.shape returns a python tuple with an entry for each dimension
m = x_train.shape[0]  # x_train.shape[0] is the length of the array and number of examples
print(f"Number of training examples is: {m}")
```

```python
# Alternative way of finding the number of training examples
m = len(x_train)
print(f"Number of training examples is: {m}")
```

```python
# Extracting various training examples from the dataset
i = 0 # Altering the value of (i) can let us view different training examples from the dataset 

x_i = x_train[i]
y_i = y_train[i]
print(f"(x^({i}), y^({i})) = ({x_i}, {y_i})")
```

```python
# Plotting the data 

# Plotting the data points
plt.scatter(x_train, y_train, marker='x', c='r')
# Setting the title
plt.title("Housing Prices")
# Setting the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
# Setting the x-axis label
plt.xlabel('Size (1000 sqft)')
plt.show()
```

```python
# Initializing the value of parameters 
w = 200
b = 100
print(f"w: {w}")
print(f"b: {b}")
```

```python
# Function to compute the models output for our training data
def compute_model_output(x_train, w, b):
    """
    Computes the prediction of a linear model
    Args:
      x (ndarray (m,)): Data, m examples 
      w,b (scalar)    : model parameters  
    Returns
      f_wb (ndarray (m,)): model prediction
    """
    m = x_train.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x_train[i] + b
        
    return f_wb
```

```python
# Computing the models output for our training data
tmp_f_wb = compute_model_output(x_train, w, b,)

# Plot our model prediction
plt.plot(x_train, tmp_f_wb, c='b',label='Our Prediction')

# Plot the data points
plt.scatter(x_train, y_train, marker='x', c='r',label='Actual Values')

# Set the title
plt.title("Housing Prices")
# Set the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
# Set the x-axis label
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```

```python
# Predicting the models output on unseen data
w = 200                         
b = 100    
x_i = 1.2  # A house with 1200 sqft. as a feature
cost_1200sqft = w * x_i + b    

print(f"${cost_1200sqft:.0f} thousand dollars")
```
