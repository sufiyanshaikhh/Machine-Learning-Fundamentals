## Linear Regression : Cost Function 

Implementing Cost Function J(w, b) for LR with one variable

```python
# Importing dependencies
import numpy as np  # For scientific computation
import matplotlib.pyplot as plt  # For plotting data
```

```python
# Creating dataset for house price prediction. The dataset constitutes of two points wherein the units for size of the house is in 1000s of sqft.
# and price of the house is in 1000s of dollars
x_train = np.array([1.0, 2.0])           #(size in 1000s of square feet)
y_train = np.array([300.0, 500.0])           #(price in 1000s of dollars)
```

```python
# Computing Cost by looping across each training point in the dataset
def compute_cost(x, y, w, b): 
    """
    Computes the cost function for linear regression.
    
    Args:
      x (ndarray (m,)): Data, m examples 
      y (ndarray (m,)): target values
      w,b (scalar)    : model parameters  
    
    Returns
        total_cost (float): The cost of using w,b as the parameters for linear regression
               to fit the data points in x and y
    """
    # number of training examples
    m = x.shape[0] 
    
    cost_sum = 0  # The total cost is summed over
    for i in range(m): 
        f_wb = w * x[i] + b  # Models prediction is computed
        cost = (f_wb - y[i]) ** 2  # The difference between the target and prediction is computed and squared
        cost_sum = cost_sum + cost  # The computed cost for corresponding training point is added to the total cost 
    total_cost = (1 / (2 * m)) * cost_sum  

    return total_cost
```

```python
# Creating a dataset on house price prediction for testing
x_train = np.array([1.0, 1.7, 2.0, 2.5, 3.0, 3.2])
y_train = np.array([250, 300, 480,  430,   630, 730,])
```

```python
# Initializing parameters w and b
w = 200
b = 0
```

```python
# Function to compute the models output for our training data
def compute_model_output(x_train, w, b):
    """
    Computes the prediction of a linear model
    Args:
      x (ndarray (m,)): Data, m examples 
      w,b (scalar)    : model parameters  
    Returns
      f_wb (ndarray (m,)): model prediction
    """
    m = x_train.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x_train[i] + b
        
    return f_wb
```

```python
# Plotting the data 

# Plotting the data points
plt.scatter(x_train, y_train, marker='x', c='r')
# Setting the title
plt.title("Housing Prices")
# Setting the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
# Setting the x-axis label
plt.xlabel('Size (1000 sqft)')
plt.show()
```

```python
# Computing the models output for our training data
tmp_f_wb = compute_model_output(x_train, w, b,)

# Plot our model prediction
plt.plot(x_train, tmp_f_wb, c='b',label='Our Prediction')

# Plot the data points
plt.scatter(x_train, y_train, marker='x', c='r',label='Actual Values')

# Set the title
plt.title("Housing Prices")
# Set the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
# Set the x-axis label
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```

```python
# Computing Cost Function J(w, b) for chosen parameters
compute_cost(x_train, y_train, w, b)
```
