## Classification

Understanding the need of a classification algorithm. This is achieved by understanding regression on categorical data and why it fails on categorical data. [contrasting regression and classification]

```python
# Importing dependencies
import numpy as np  # For scientific computation
import matplotlib.pyplot as plt  # For plotting data
from lab_utils_common import dlc, plot_data
```

```python
# Creating a dataset to understand classification problems
x_train = np.array([0., 1, 2, 3, 4, 5])
y_train = np.array([0,  0, 0, 1, 1, 1])
X_train2 = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y_train2 = np.array([0, 0, 0, 1, 1, 1])
```

```python
pos = y_train == 1
neg = y_train == 0

fig,ax = plt.subplots(1,2,figsize=(8,3))
#plot 1, single variable
ax[0].scatter(x_train[pos], y_train[pos], marker='x', s=80, c = 'red', label="y=1")
ax[0].scatter(x_train[neg], y_train[neg], marker='o', s=100, label="y=0", facecolors='none', 
              edgecolors=dlc["dlblue"],lw=3)

ax[0].set_ylim(-0.08,1.1)
ax[0].set_ylabel('y', fontsize=12)
ax[0].set_xlabel('x', fontsize=12)
ax[0].set_title('one variable plot')
ax[0].legend()

#plot 2, two variables
plot_data(X_train2, y_train2, ax[1])
ax[1].axis([0, 4, 0, 4])
ax[1].set_ylabel('$x_1$', fontsize=12)
ax[1].set_xlabel('$x_0$', fontsize=12)
ax[1].set_title('two variable plot')
ax[1].legend()
plt.tight_layout()
plt.show()
```

In the single variable plot, positive results are shown both a red 'X's and as y=1. Negative results are blue 'O's and are located at y=0. In the case of linear regression, y would not have been limited to two values but could have been any value.

In the two-variable plot, the y axis is not available. Positive results are shown as red 'X's, while negative results use the blue 'O' symbol. In the case of linear regression with multiple variables, y would not have been limited to two values and a similar plot would have been three-dimensional.

## Linear Regression on Categorical data

For some scenarios, linear regression may seem to adapt well to the training data, but when observed on several other occassions with outliers, the model may deflect and presumes incorrect predictions for some of the input features. This is mainly because linear regression averages the cost between the actual value and the predicted value across all the training examples. 

The threshold occasionally shifts when the training data contains outliers leading to incorrect predictions on several occassions. This clearly indicates the need of a seperate algorithm to perform prediction on categorical data. There comes the classification algorithm namely logistic regression into the picture.

The classification algorithm forms a decision boundary that seperates two or more classes in the training data. This is mainly referred to as z in mathematical function and given out as z = w.x + b. This z if further passed on to a sigmoid function that predicts the outcome in a range of 0 to 1. The outcome is further evaluated by a threshold value and corresponding class is predicted.
