## Decision Boundary for Logistic Regression

Here, we will be plotting a decision boundary for a logistic regression model. This will give us a better sense on how the model is making predictions.

```python
# Importing the dependencies
import numpy as np  # For scientific computations
import matplotlib.pyplot as plt  # For plotting data
from utils import plot_data, sigmoid, draw_vthresh
```

```python
# Creating a dataset wherein there are 6 training examples stored in a numpy array where each example inputs 2 features.
# The classification of these training examples is stored in a numpy array y.
X = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y = np.array([0, 0, 0, 1, 1, 1]).reshape(-1,1)
```

```python
# Plotting the data for visualization
fig,ax = plt.subplots(1,1,figsize=(4,4))
plot_data(X, y, ax)

ax.axis([0, 4, 0, 3.5])
ax.set_ylabel('$x_1$')
ax.set_xlabel('$x_0$')
plt.show()
```

The Red crosses demonstrates positive classification (y = 1) whereas the blue circles are negative classifications (y = 0)

## Logistic Regression and Decision Boundary

The output of the logistic regression model f(x) is interepreted as the probablity that y = 1 given x, and parameterized w and b. Therefore, to get a final prediction of the logistic regression model, we can use the heuristic, if the sigmoid function g(z) outputs a value >= 0.5, the model predicts y = 1 and if the sigmoid function (z) outputs a value < 0.5, the model predicts y = 0.

```python
# Plot sigmoid(z) over a range of values from -10 to 10
z = np.arange(-10,11)

fig,ax = plt.subplots(1,1,figsize=(5,3))
# Plot z vs sigmoid(z)
ax.plot(z, sigmoid(z), c="b")

ax.set_title("Sigmoid function")
ax.set_ylabel('sigmoid(z)')
ax.set_xlabel('z')
draw_vthresh(ax,0)
```

We could clearly see that the sigmoid function g(z) outputs a value >= 0.5 when the value of z is greater than 0 and for z to be greater than 0, w.x + b should be greater than 0 (where parameter w is a vector and b is scalar)

So, the whole idea comes down to, when w.x + b >= 0, the model predicts y = 1, and when w.x + b < 0, the model predicts y = 0

## Plotting the decision boundary

Now, it is pretty much clear that how logistic regression model is predicting classes. The logistic regression model has the form f(x) = g(-3 + x0 + x1) where the parameters are intialized manually for intuition.

We now know that this model predicts the positive class when -3 + x0 + x1 >= 0. To plot the decision boundary, we need to equate this equation by 0 which is -3 + x0 + x1 = 0, which is equivalent to x1 = 3 âˆ’ x0.

```python
# Choosibg values between 0 and 6
x0 = np.arange(0,6)

x1 = 3 - x0
fig,ax = plt.subplots(1,1,figsize=(5,4))
# Plot the decision boundary
ax.plot(x0,x1, c="b")
ax.axis([0, 4, 0, 3.5])

# Fill the region below the line
ax.fill_between(x0,x1, alpha=0.2)

# Plot the original data
plot_data(X,y,ax)
ax.set_ylabel(r'$x_1$')
ax.set_xlabel(r'$x_0$')
plt.show()
```

In the plot above, the blue line represents the line x0 + x1 âˆ’ 3 = 0 and it should intersect the x1 axis at 3 (if we set x1 = 3, x0 = 0) and the x0 axis at 3 (if we set x1 = 0, x0 = 3). The shaded region represents âˆ’3 + x0 + x1 < 0. The region above the line is âˆ’3 + x0 + x1 > 0.

Any point in the shaded region (under the line) is classified as y = 0 . Any point on or above the line is classified as y = 1. This line is known as the "decision boundary".

By using higher order polynomial terms, we come up with more complex decision boundaries.
